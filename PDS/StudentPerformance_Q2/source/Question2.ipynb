```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/cleaned_student.csv")
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='test preparation course', data=data)
plt.title('Count Plot of Test Preparation Course Completion')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.show()
```

```python
#2 Bar Plot of Average Writing Scores by Race/Ethnicity
plt.figure(figsize=(10, 6))
sns.barplot(x='race/ethnicity', y='writing score', data=data, ci=None)
plt.title('Average Writing Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Writing Score')
plt.show()
```

```python
#3: Scatter Plot of Math vs. Reading Scores
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', data=data)
plt.title('Scatter Plot of Math vs. Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

```python
#4: Box Plot of Reading Scores by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='gender', y='reading score', data=data)
plt.title('Box Plot of Reading Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Reading Score')
plt.show()
```

```python
#5: Histogram of Math Scores
plt.figure(figsize=(8, 6))
sns.histplot(data['math score'], bins=20, kde=True)
plt.title('Histogram of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```
